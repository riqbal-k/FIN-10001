```python
import pandas as pd
import numpy as np
from scipy import stats
```

# Upload and Load the Dataset

```python
df = pd.read_csv("/kaggle/input/spydata/week 11 Tutorial data.csv")
```

```python
#Head (first rows of the DataFrame)
df.head()
```

```python
#Check what your Dates column looks like for df
print(df['Date'].dtype)
```

```python
#Convert Dates to datetime format
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
# Calculate log returns
df['spy_log_return'] = np.log(df['SPY'] / df['SPY'].shift(1))
```

```python
#Head (first rows of the DataFrame)
df.head()
```

```python
# Drop the first row because it will be NaN
df = df.dropna()

# Display the result
df.head()
```

# Confidence Interval Function

```python
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)              # standard error of the mean
    h = sem * stats.t.ppf((1 + confidence) / 2, n - 1)   # margin of error
    lower = mean - h
    upper = mean + h
    return mean, lower, upper, h
```

# Confidence intervals for monthly returns (spy_log_return)

```python
returns = df["spy_log_return"]
```

```python
mean_82, lower_82, upper_82, moe_82 = confidence_interval(returns, confidence=0.82)
mean_79, lower_79, upper_79, moe_79 = confidence_interval(returns, confidence=0.79)
```

# Confidence interval for monthly prices (SPY)

```python
prices = df["SPY"]

mean_93, lower_93, upper_93, moe_93 = confidence_interval(prices, confidence=0.93)
```

# PRINT RESULTS

```python
print("82% Confidence Interval for Monthly Returns:")
print("Lower:", lower_82)
print("Upper:", upper_82)
print("Margin of Error:", moe_82)
print()
```

```python
print("79% Confidence Interval for Monthly Returns:")
print("Lower:", lower_79)
print("Upper:", upper_79)
print("Margin of Error:", moe_79)
print()
```
